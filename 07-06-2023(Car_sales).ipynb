## <center><u><b>PLOTING WITH PANDAS

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('./Car_sales.csv')
data
```

#### <i>Dropping Nan Values

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
data.rename(columns={"__year_resale_value" :"year_resale_value"},inplace=True)
```

```python
data.head()
```

### <i><b>Plots on numerical value columns

### <i>Histogram

```python
data.Horsepower.plot.hist(cmap="autumn");
```

### <i>BoxPlot

```python
data.Sales_in_thousands.plot.box(figsize=(10,6),cmap="twilight_shifted");
```

```python
data.Horsepower.plot.box(figsize=(4,3),cmap="rainbow");
```

```python
data.Price_in_thousands.plot.box(figsize=(6,3),cmap="tab20");
```

### <i>Kernel Density Estimation Plot(KDE)

```python
data.Sales_in_thousands.plot.kde();  
```

```python
data.Fuel_capacity.plot.kde(cmap="Wistia");
```

### <i>Subplot

```python
data.plot.box(figsize=(23,5),subplots=True);
```

```python
data.plot.hist(figsize=(6,18),subplots=True);
```

```python
data.plot.kde(figsize=(7,20),subplots=True);
```

```python
data.plot.kde(figsize=(10,3.7));
```

### <i>Linechart

```python
data.sort_values(by="year_resale_value").plot.line("year_resale_value","Price_in_thousands",figsize=(6,3),cmap="inferno");
```

```python
data.plot.line(x="Manufacturer",y="Fuel_capacity",cmap="rainbow")
```

### <i>Bar Plot
#### To create the below barchart we have to create the frequency table.Frequency table can be using the value_counts()method

```python
data.Manufacturer.value_counts().plot.bar(figsize=(6,2),cmap="turbo_r");
```

```python
data.Model.value_counts().plot.bar(figsize=(30,9),cmap="Purples_r");
```

```python
data.sort_values(by="Price_in_thousands")[:20].plot.bar("Manufacturer","Price_in_thousands",figsize=(10,3),cmap="twilight");
```

```python
data.sort_values(by="Engine_size")[:10].plot.bar("Model","Engine_size",figsize=(9,3),cmap="Paired");
```

### <i>Area Chart

```python
data.plot.area("Manufacturer","Price_in_thousands",figsize=(9,3),cmap="Pastel1");
```

### <i>Horizontal bar chart

```python
data.sort_values(by="Price_in_thousands")[:5].plot.barh("Model","Price_in_thousands",figsize=(10,3),cmap="Pastel2");
```

### <i>Scatter Plot

```python
data.sort_values(by="Price_in_thousands")[:8].plot.scatter("Latest_Launch","Price_in_thousands",figsize=(10,4));
```

```python
data.sort_values(by="Price_in_thousands")[-15:].plot.scatter("Latest_Launch","Price_in_thousands",figsize=(18,4));
```

### <i>Pie Chart

```python
data.Vehicle_type.value_counts().plot.pie();
```

```python
data.Vehicle_type.value_counts().plot.pie(autopct="%1.3f%%");
```

```python
### Box plot
data.plot.box(figsize=(22,5));
```

```python
data.sort_values(by="Engine_size",ascending=False)[:10].Engine_size.plot.pie();
```

### <i>Hexbin Plot

```python
data.plot.hexbin("Fuel_efficiency","Curb_weight",gridsize=(10,5),cmap="afmhot");
```

```python
data.plot.hexbin("Fuel_capacity","Engine_size", gridsize=20);
```
