## <center><u><b> Extracting the Date

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('./Car_sales.csv',parse_dates=["Latest_Launch"])
data
```

#### <i> Changed Latest_Launch to datetime64[ns] format

```python
data.info()
```

### <u>Sorting 

```python
data.sort_values(by="Latest_Launch",inplace=True)
data
```

#####  Dataset sorted according to Date(Latest_Launch)

### <u> Resetting Index

```python
data.reset_index(drop=True,inplace=True)
```

```python
data.head(10)
```

### <u>Feature Engineering

```python
data["Year"]=data.Latest_Launch.dt.year
data["Month"]=data.Latest_Launch.dt.month
data["Day"]=data.Latest_Launch.dt.day
```

```python
data
```

##### New Columns called Year,Month,Day is created 

### <u>Renaming column

```python
data.rename(columns={"__year_resale_value":'year_resale_value'},inplace=True)
```
