## <u><center>Plotting with Matplotlib and Seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('./Car_sales.csv',parse_dates=['Latest_Launch'])
data
```

```python
data.sort_values(by="Latest_Launch",inplace=True)
data
```

```python
data.info()
```

```python
data.reset_index(drop=True,inplace=True)
data
```

```python
data.describe()
```

```python
data.isnull().mean()*100
```

```python
data.dropna(inplace=True)
data.drop_duplicates(inplace=True)
```

```python
data["Year"]=data.Latest_Launch.dt.year
data["Month"]=data.Latest_Launch.dt.month
data["Day"]=data.Latest_Launch.dt.day
data.head(3)
```

```python
data.drop(columns="Latest_Launch",inplace=True)
```

```python
data.shape
```

```python
data.info()
```

### <u>Univariate Analysis

#### Histogram

```python
sns.histplot(data=data,x="Sales_in_thousands")
plt.show()
```

#### Boxplot

```python
sns.boxplot(data=data,y="Sales_in_thousands")
plt.show()
```

#### Kernel Density Estimation Plot(KDE)

```python
sns.kdeplot(data=data.Sales_in_thousands)
plt.show()
```

#### <u>Subplot

```python
# Subplot of histogram
sns.histplot(data=data)
plt.show()
```

```python

plt.figure(figsize=(14,6))
plt.xticks(rotation=95)
sns.boxplot(data=data)
plt.show()
```

```python
sns.kdeplot(data=data)
plt.show()
```

### <u><b>For Loop to plot subplot

```python
#boxplot

for i in data.columns:
    if data[i].dtypes !="object":
        sns.boxplot(y=data[i])
    plt.show()
```

```python
# histogram
for i in data.columns:
    if data[i].dtypes !='object':
        sns.histplot(x=data[i])   
    plt.show()
```

```python
# kdeplot
for i in data.columns:
    if data[i].dtypes !='object':
        sns.kdeplot(data=data[i])
    plt.show()
```

### <u> Subplots in matplotlib

```python
plt.figure(figsize=(10,4))
plt.subplot(2,2,1)
sns.boxplot(data=data,y="Sales_in_thousands")

plt.figure(figsize=(10,4))
plt.subplot(2,2,2)
sns.boxplot(data=data,y='Price_in_thousands')

plt.figure(figsize=(10,4))
plt.subplot(2,2,3)
sns.boxplot(data=data,y='Engine_size')

plt.figure(figsize=(10,4))
plt.subplot(2,2,4)
sns.boxplot(data=data,y='Horsepower')

plt.show()
```

```python
#subplot of histogram
plt.figure(figsize=(10,4))
plt.subplot(2,2,1)
sns.histplot(data=data,y="Sales_in_thousands")

plt.subplot(2,2,2)
sns.histplot(data=data,y='Price_in_thousands')

plt.subplot(2,2,3)
sns.histplot(data=data,y='Engine_size')

plt.subplot(2,2,4)
sns.histplot(data=data,y='Horsepower')

plt.show()
```

```python
# subplot of kde
plt.figure(figsize=(10,5))
plt.subplot(2,2,1)
sns.kdeplot(data=data,y='Horsepower')

plt.subplot(2,2,2)
sns.kdeplot(data=data,y='Price_in_thousands')

plt.subplot(2,2,3)
sns.kdeplot(data=data,y='Engine_size')

plt.subplot(2,2,4)
sns.kdeplot(data=data,y='Sales_in_thousands')

plt.show()
```

### <u>Subplot using Enumerate


```python
feature= []
```

```python
for i in data.columns:
    if data[i].dtypes !='object':
        feature.append(i)
```

```python
feature
```

```python
list(enumerate(feature))
```

```python
#boxplot
plt.figure(figsize=(13,10))
for i in enumerate(feature):
    plt.subplot(5,3,i[0]+1)
    sns.boxplot(data=data,y=i[1],color="green")
```

### Line charts

```python
plt.figure(figsize=(10,6))
plt.xlabel=("Year resale value")
plt.ylabel=('Price')
plt.title("Car_sales")
sns.lineplot(data=data,x='Year_resale_value',y='Price_in_thousands')
plt.show()
```

### Countplot
