##### Adding new  column from Date

```python
data['Year']=data.Date.dt.year
data['Month']=data.Date.dt.month
data['weekday']=data.Date.dt.day
data
```

```python
data['Month']=data.Month.replace({1:'January',
                                2:'February',
                                3:'March',
                                4:'April',
                                5:'May',
                                6:'June',
                                7:'July',
                                8:'August',
                                9:'September',
                                10:'October',
                                11:'November',
                                12:'December'}
                                )
data['weekday']=data.weekday.replace({0:'Sunday',
                                   1:'Monday',
                                   2:'Tuesday',
                                   3:'Wednesday',
                                   4:'Thursday',
                                   5:'Friday',
                                   6:'Saturday'})
data.head(10)
```

#### Temperature conversion(-2&#176; -32)X 5/9

```python
data['Celsius']=[(Temp-32)*(5/9) for Temp in data.Temperature]
```

#### Create a new column called Heat

```python
data['Heat']=pd.cut(x=data.Celsius,bins=[-28,10,35],labels=['cold','warm'])
data
```

```python
sns.barplot(x='Heat',y="Weekly_Sales",hue='Holiday_Flag',data=data)
plt.legend(bbox_to_anchor=[1.2,1])
plt.show()
```
