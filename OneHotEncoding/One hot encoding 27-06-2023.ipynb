## <center> One Hot Encoding and Feature Scaling

```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

### Loading Dataset

```python
data = sns.load_dataset('titanic')
```

```python
data
```

```python
data.dropna(inplace = True)
data.drop_duplicates(inplace = True)
```

```python
data
```

### One hot encoding

```python
data['SEX'] = pd.get_dummies(data= data.sex , drop_first= True)
```

```python
data.head(5)
```

### Converting the entire dataset

```python
data_dummies = pd.get_dummies(data=data)
```

```python
data_dummies
```

```python
data_dummies.shape
```

```python
data_with_dummies = pd.get_dummies(data= data,drop_first= True)
data_with_dummies
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler,RobustScaler,LabelEncoder
```

### Label Encoder

```python
le = LabelEncoder()
```

```python
data['EMBARKED'] = le.fit_transform(data.embarked)
data
```

## <u>Data Scaling and Data Normalising

### <li> MinMax Scaling

```python
minmax = MinMaxScaler()
```

```python
data[['age']]
```

```python
age = data[['age']].values
```

```python
data["AGE"] = minmax.fit_transform(age)
data.head(4)
```

```python
data.AGE.min()
```

```python
data.AGE.max()
```

### <li>Standard Scaler

```python
sc = StandardScaler()
```

```python
data['AGE_SC'] = sc.fit_transform(age)
data
```

```python
data.AGE_SC.min()
```

```python
data.AGE_SC.max()
```

```python
data['FARE_SC'] = sc.fit_transform(data[['fare']].values)
data
```
