### Dropping Row Id Column and Postal Code

```python
data.drop(columns=['Row ID','Postal Code'],inplace=True)
data.head(3)
```

## Renaming Columns

```python
data.rename(columns={'Order ID':'Order_ID','Order Date':'OrderDate','Ship Date':'ShipDate','Ship Mode':'ShipMode','Customer Name':'CustomerName',
                     'Product ID':'ProductID','Product Name':'ProductName','Sub-Category':'Sub_Category'},inplace=True)
data.sample()
```

### Creating a New column "Delay" using ShipDate and OrderDate

```python
data['Delay']=data['ShipDate']-data['OrderDate']
data['Delay']
```

```python
data.info()
```

### Changing Delay Column datatype to string

```python
data["Delay"]=data['Delay'].astype('str')
```

```python
data["Delay1"]=data['Delay'].str.replace("\D"," ",regex=True)
data["Delay1"]=data["Delay1"].astype('int')
data["Delay1"]
```

### Creating a Column Weekday from Delay Column

```python
data["Weekday"]=data.OrderDate.dt.day_of_week
data.Weekday.unique()
```

### Replacing Weekday from number format to Days 

```python
data.Weekday.replace({0:'Sunday',1:'Monday',2:'Tuesday',3:'Wednesday',4:'Thursday',5:'Friday',6:'Saturday'},inplace=True)
```

```python
data.head()
```

### Dropping Delay Column

```python
data.drop(columns="Delay",inplace=True)
```
